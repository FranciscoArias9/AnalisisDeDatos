```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("/content/casasboston.csv")

df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]

df = datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"

})

print(df)

df.N_HABITACIONES_MEDIO.plot.hist(color="lightblue")
df.N_HABITACIONES_MEDIO.plot.hist(title="numero de habitaciones medio")

#df.N_HABITACIONES_MEDIO.set_facecolor('lightblue')
plt.show()


import numpy as np

x = np.arange(0, 20, 0.05)
y = np.sin(x*np.pi)

fig, ax = plt.subplots(figsize=(3,2), constrained_layout=True)
ax.plot(x, y)
ax.set_xlabel('y [s]')
ax.set_ylabel('x [V]')
ax.set_title('green')
fig.set_facecolor('lightblue')




```

```python
df.plot.scatter(x="VALOR_MEDIANO",y="INDICE_CRIMEN",alpha=0.5,color="red")
plt.show()
```
